```julia
## Plots on Julia
## daniel-rs

## Paquetes
using Plots
using DataFrames
using RDatasets
using StatsPlots
```

```julia
x = rand(1000, 2);
```

```julia
@time scatter(x[:,1],x[:,2])
```

```julia
y = randn(1000,2);
```

```julia
@time scatter(y[:,1], y[:,2], m = 0.5)
```

```julia
f(x) = sin(2x)^3 
```

```julia
plot([sin cos f], -2π, 2π, lw = [2 2 3], label = ["sin" "cos" "sin(2x)^3"])
title!("Funciones Trigonometricas")
plot!(legend=:outerbottom, legendcolumns=3)
```
