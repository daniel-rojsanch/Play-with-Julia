```julia
using Distributions
using Plots
using StatsPlots
```

```julia
## crear observaciones pseudoaleatorias que provengan de una variable aleatoria Normal

x = rand(Normal(), 1000);
histogram(x, normalize = true, label = "Obs", c = :blue4)
plot!(Normal(), lw = 3, label = "Teoric", c = :orange)
```

```julia
mean(Normal())
```

```julia
mean(x)
```

```julia
var(Normal())
```

```julia
var(x)
```

```julia
std(x)
```

```julia
fit(Normal,x)
```

```julia
y = rand(Beta(4, 15), 1000);
```

```julia
histogram(y, normalize = true)
plot!(Beta(4,15))
```

```julia
fit(Beta, y)
```

```julia
scatter(x, y, label = "Obs", alpha = 0.8, c = :tomato)
```
